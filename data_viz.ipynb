```python
import plotly
import plotly.express as px
import pandas as pd
```

```python
researchers_data = pd.read_csv('researchers_data.csv')  
```

```python
researchers_data
```

```python
#get parent vals
parents = researchers_data["researcher_type_str"].unique()
parents
```

```python
df = px.data.tips()
df
```

```python
#plot of how many projects are being worked on in each research interest field
fig1 = px.pie(researchers_data, values='project_working_on', names='research_interest')
fig1.show()
```

```python
#plot breakdown of project interests by researcher type
fig2 = px.bar(researchers_data, x="research_interest", y="project_working_on", color="researcher_type_str", 
             title="Breakdown of Project Interests by Researcher Type")
fig2.show()
```
